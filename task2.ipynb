```python
import sys
!{sys.executable} -m pip install umap-learn
!{sys.executable} -m pip install scikit-learn
!{sys.executable} -m pip install --force-reinstall numpy==1.19
from umap import UMAP
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.datasets import load_digits
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
```

```python
umap = UMAP(random_state=2)
tsne = TSNE(n_jobs=-1, random_state=2)
pca = PCA(n_components=2, random_state=2)
```

```python
digits = load_digits()
mnist = pd.DataFrame(digits.data)
fmnist = pd.read_csv('fmnist.csv')
word_vectors = []
coil20 = []
```

```python
models = {"UMAP": umap, "PCA": pca, "t-SNE": tsne}
# datasets = {'MNIST': mnist, "Fashion MNIST": fmnist}
```

```python
embedding = pca.fit_transform(mnist)
plt.scatter(embedding[:, 0], embedding[:, 1], c=digits.target, cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
plt.colorbar(boundaries=np.arange(11)-0.5).set_ticks(np.arange(10))
plt.title("PCA");
plt.show()
```

```python
embedding = tsne.fit_transform(mnist)
plt.scatter(embedding[:, 0], embedding[:, 1], c=digits.target, cmap='Spectral', s=5)
plt.gca().set_aspect('equal', 'datalim')
plt.colorbar(boundaries=np.arange(11)-0.5).set_ticks(np.arange(10))
plt.title("t-SNE");
plt.show()
```
