```python
import pandas as pd
from umap import UMAP
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
mnist_url = "https://www.openml.org/data/get_csv/52667/mnist_784.arff"
mnist_df = pd.read_csv(mnist_url)
```

```python
plt.figure(figsize=(10, 10))
print("Алгоритм UMAP на MNIST")
MNIST_e = UMAP().fit_transform(mnist_df[:20000].drop('class', axis=1))
plt.scatter(MNIST_e[:,0], MNIST_e[:,1], c=mnist_df[:20000]['class'], cmap="plasma", s=2)

plt.show()
```

```python
print("Алгоритм TSNE на MNIST")
plt.figure(figsize=(10, 10))
MNIST_e = TSNE(n_components=2).fit_transform(mnist_df[:5000].drop('class', axis=1))
# Памяти хватило только на 5000
plt.scatter(MNIST_e[:,0], MNIST_e[:,1], c=mnist_df[:5000]['class'], cmap="plasma", s=2)
plt.show()
```

```python
print("Алгоритм PCA на MNIST")
plt.figure(figsize=(10, 10))
MNIST_e = PCA(n_components=2).fit_transform(mnist_df[:20000].drop('class', axis=1))
plt.scatter(MNIST_e[:,0], MNIST_e[:,1], c=mnist_df[:20000]['class'], cmap="plasma", s=2)
plt.show()
```

```python
coil_20_df = pd.read_csv('https://raw.githubusercontent.com/vaksakalli/datasets/master/coil20.csv')
```

```python
plt.figure(figsize=(10, 10))
print("Алгоритм UMAP на COIL20")
coil_e = UMAP().fit_transform(coil_20_df.drop('1024', axis=1))
plt.scatter(coil_e [:,0], coil_e [:,1], c=coil_20_df['1024'], cmap="plasma", s=2)

plt.show()
```

```python
plt.figure(figsize=(10, 10))
print("Алгоритм TSNE на COIL20")
coil_e = TSNE(n_components=2).fit_transform(coil_20_df.drop('1024', axis=1))
plt.scatter(coil_e [:,0], coil_e [:,1], c=coil_20_df['1024'], cmap="plasma", s=2)

plt.show()
```

```python
plt.figure(figsize=(10, 10))
print("Алгоритм PCA на COIL20")
coil_e = PCA(n_components=2).fit_transform(coil_20_df.drop('1024', axis=1))
plt.scatter(coil_e [:,0], coil_e [:,1], c=coil_20_df['1024'], cmap="plasma", s=2)

plt.show()
```

```python
__,__, f_mnist_x_test, f_mnist_y_test = dataget.image.fashion_mnist().get()
f_mnist = f_mnist_x_test.reshape(-1, 28*28)
```

```python
pca = PCA(n_components=2)
print("Алгоритм PCA на FMNIST")
embedding_pca_f_mnist = pca.fit_transform(f_mnist)

plt.figure(figsize=(10, 10))
plt.scatter(embedding_pca_f_mnist[:,0], embedding_pca_f_mnist[:,1], c = f_mnist_y_test, cmap = 'Spectral', s = 5)
plt.show()
```

```python
tsne = TSNE(n_jobs=-1, random_state=2)
print("Алгоритм TSNE на FMNIST")
embedding_tsne_f_mnist = tsne.fit_transform(f_mnist)

plt.figure(figsize=(10, 10))
plt.scatter(embedding_tsne_f_mnist[:,0], embedding_tsne_f_mnist[:,1], c = f_mnist_y_test, cmap = 'Spectral', s = 2)
plt.show()
```

```python
umap = UMAP(random_state = 2)
print("Алгоритм UMAP на FMNIST")
embedding_umap_f_mnist = umap.fit_transform(f_mnist)

plt.figure(figsize=(10, 10))
plt.scatter(embedding_umap_f_mnist[:,0], embedding_umap_f_mnist[:,1], c = f_mnist_y_test, cmap = 'Spectral', s = 2)
plt.show()
```

```python
from google.colab import drive
drive.mount('/content/drive')
from gensim import models

!wget -c "https://s3.amazonaws.com/dl4j-distribution/GoogleNews-vectors-negative300.bin.gz"

!gzip -d GoogleNews-vectors-negative300.bin.gz

w = models.KeyedVectors.load_word2vec_format(
    '/content/GoogleNews-vectors-negative300.bin', binary=True)

w_vectors = w.wv.vectors
```

```python
w_vectors = w_vectors[np.random.randint(w_vectors.shape[0], 
                size=int(w_vectors.shape[0]/60)), :]
print(w_vectors.shape)
```

```python
pca = PCA(n_components=2)
print("Алгоритм PCA на Google news")
embedding_pca_G_news = pca.fit_transform(w_vectors)
plt.figure(figsize=(10, 10))
plt.scatter(embedding_pca_G_news[:,0], embedding_pca_G_news[:,1], 
                         cmap = 'Spectral', s = 1)
plt.show()
```

```python
umap = UMAP(random_state = 2)
print("Алгоритм UMAP на Google news")
embedding_umap_G_news = umap.fit_transform(w_vectors)
plt.figure(figsize=(10, 10))
plt.scatter(embedding_umap_G_news[:,0], embedding_umap_G_news[:,1], 
                         cmap = 'Spectral', s = 1)
plt.show()
```

```python
tsne = TSNE(n_jobs=-1, random_state=2)
print("Алгоритм TSNE на Google news")
embedding_tsne_G_news = tsne.fit_transform(w_vectors)
plt.figure(figsize=(10, 10))
plt.scatter(embedding_tsne_G_news[:,0], embedding_tsne_G_news[:,1], 
                         cmap = 'Spectral', s = 1)
plt.show()
```
